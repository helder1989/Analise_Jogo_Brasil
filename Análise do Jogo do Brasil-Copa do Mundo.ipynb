Incluindo os confrontos em copa, na história foram cinco jogos (contando com o último jogo realizado em 09/12/2022).

## Importando as Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://fbref.com/pt/partidas/bb30147d/Croatia-Brazil-2022Dezembro9-World-Cup'
```

```python
match = pd.read_html(url, header=1)
```

```python
# Carregando os dados da tabela chutes, indíce número 17

shots_both_squads=match[17]
shots_both_squads
```

## Preparando os dados(xG)

```python
shots_both_squads.info()
```

## Ajustando alguns dados

Ao lado do nome dos times (coluna 'Equipe") consta um br"Brasil" e hr "Croácia", letras mínuscula e espaço no qual representa a bandeira de cada equipe, sendo assim será feito um ajuste, pois não vai ficar interessante no gráfico. 

```python
shots_both_squads['Equipe'] = shots_both_squads['Equipe'].str.slice(start=3)

#queremos o valor a partir da terceira letra, sendo assim não será considerado br ou hr e o espaço. 
```

```python
shots_both_squads.head()
```

## Estipular que os valores xG da equipe visitante seja negativo para melhorar visualização

```python
away_team = 'Brazil'
shots_both_squads['xG'] = np.where(shots_both_squads['Equipe'] == away_team, shots_both_squads['xG'] * -1, shots_both_squads['xG'])
```

O valor de um time vai ficar pra cima e o outro pra baixo, aplicando uma função do **numpy, chamada where** que vai aonde a equipe foi igual ao **away_team** for igual à Brazil, pegando o xG multiplicando por -1, sendo ao contrário mantém o mesmo. 

```python
shots_both_squads.head()
```

## Gráficos xG

Apresentando ao longo da partida, conforme os minutos passam como foi a qualidade dos chutes de cada equipe.

```python
plt.figure(figsize=(10, 7))
plt.title('Gols Esperados ao longo da partida', fontweight='bold', fontsize='14')

sns.set_theme('paper')
sns.barplot(data=shots_both_squads, x='Minute', y='xG', hue='Equipe')
plt.legend(loc='upper right');
```

Através do gráfico de barras, que a equipe do Brazil realizou mais chutes do que a Croácia. Lembrando que essas linhas pretas ao final de algumas barras significa que houve mais chutes naquele minuto. 

## Preparação dos dados(Passes)

```python
brazil = match[11]
brazil['Equipe'] = 'Brazil' #criando uma coluna com o nome da equipe
brazil = brazil[:-1] #exluindo a última linda da coluna que é a somatória de todas e não queremos
brazil.tail() #as cinco últimas linhas
```

```python
croacia = match[4]
croacia['Equipe'] = 'Croatia'#criando uma coluna com o nome da equipe
croacia = croacia[:-1] #exluindo a última linda da coluna que é a somatória de todas e não queremos
croacia.tail() #as cinco últimas linhas
```

```python
#concatenação das duas equipes

teams = pd.concat([brazil, croacia])
teams = teams[['Equipe', '1/3']] #colunas selecionadas #quantidade de passes para o último 1/3
teams
```

## Gráfico (Passes)

```python
plt.figure(figsize=(15, 5))
plt.title('Passes para o Último Terço do Campo', fontweight='bold', fontsize=14)

sns.set_theme('paper')
sns.boxplot(data=teams, x='1/3', y='Equipe') #boxplot pra criar um gráfico de caixas
plt.ylabel('Equipe');
```

## Conclusão

Através do gráfico de caixas conseguimos paupar que a equipe brasileira realizou mais passes no último 1/3 do campo, ou seja, o Brazil foi a equipe que ficou "tocando" bola para o último 1/3 de campo da Croácia o tempo todo. O valor máximo da Croácia fica em torno de uns 35%, enquanto o Brasil ficou em 50%.

Podemos afirmar através das análises que o Brazil pressionou mais a Croácia,e que sem assistir a partida conseguimos verificar como foi, ou seja, simplificando em gráficos simples e objetivos é sem dúvidas mais viável.
